```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
df=pd.read_csv(r"C:\Users\HP\cardio_train.csv" ,sep=';' )
df=df.drop(['id'],axis=1)
df
```

# splitting dataset into x and y

```python
x=df.iloc[:,0:11]
x
```

```python
y=df.iloc[:,11:]  
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
```

# 1.Logistic Regression(71.5% Accuracy)

```python
from sklearn.linear_model import LogisticRegression
model1=LogisticRegression()
model1.fit(xtrain,ytrain)
```

```python
ans1=model1.predict(xtest)
ans1
```

```python
model1.score(xtest,ytest)
```

# 2.SVM(60.2% Accuracy)

```python
from sklearn.svm import SVC
model2=SVC()
model2.fit(xtrain,ytrain)
```

```python
ans2=model2.predict(xtest)
ans2
```

```python
model2.score(xtest,ytest)
```

# 3.Decision Tree Classifier(63.5% Accuracy)

```python
from sklearn.tree import DecisionTreeClassifier
model3=DecisionTreeClassifier()
model3.fit(xtrain,ytrain)
```

```python
ans3=model3.predict(xtest)
ans3
```

```python
model3.score(xtest,ytest)
```

# 4.Random Forest Classifier(71.9% - Highest Accuracy)

```python
from sklearn.ensemble import RandomForestClassifier
model4=RandomForestClassifier()
model4.fit(xtrain,ytrain)
```

```python
ans4=model4.predict(xtest)
ans4
```

```python
model4.score(xtest,ytest)
```

```python
from sklearn.naive_bayes import GaussianNB
model5=GaussianNB()
model5.fit(xtrain,ytrain)
```

```python
ans5=model5.predict(xtest)
ans5
```
